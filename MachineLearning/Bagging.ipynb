```python
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import BaggingClassifier
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import accuracy_score
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/persocom01/GA-lessons/master/6.02-lesson-bagging/datasets/Heart.csv')
```

```python
df.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
df.shape
```

```python
df.head()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
df = pd.get_dummies(df, columns = ['ChestPain', 'Thal', 'AHD'], drop_first=True)
```

```python
df.head()
```

```python
x = df.drop('AHD_Yes', axis=1)
y = df['AHD_Yes']
```

```python
y.value_counts(normalize=True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(x,y, random_state=42, stratify=y)
```

```python
tree = DecisionTreeClassifier()
```

```python
cross_val_score(tree,X_train, y_train).mean()
```

```python
tree.fit(X_train, y_train)
```

```python
tree.score(X_train, y_train)
```

```python
tree.score(X_test, y_test)
```

```python
X_test.index
```

```python
predictions = pd.DataFrame(index=X_test.index)
predictions
```

```python
for i in range(10):
    t = DecisionTreeClassifier()
    t.fit(X_train, y_train)
    predictions[f'Tree{i}'] = t.predict(X_test)

predictions
```

```python
probs = predictions.mean(axis='columns')
probs
```

```python
accuracy_score(y_test, (probs>.5).astype(int))
```
