```python
import numpy as np
import random
import seaborn as sns
np.random.seed(6712)
```

```python
# Here, we have calculated normally distributed values using Numpy
# numpy.random.normal(mean=12.5, standarad deviation=3.88)
random_values = [int(np.random.normal(12.5, 3.88)) for i in range(1000)]
print("The mean of random normal distribution is:", np.mean(random_values))
```

```python
# OR We can also calculate the normal distribution values by using Random module
lst = [int(random.normalvariate(12.5, 12.5/(2*1.96))) for i in range(1000)]
```

```python
values, frequencies = np.unique(random_values, return_counts = True)
```

```python
values
```

```python
frequencies
```

```python
import seaborn as sns
sns.set_style('whitegrid')
axes = sns.barplot(x = values, y = frequencies, color = 'gray')
axes.set_title('Normal Distribution')
axes.set(xlabel = 'Value', ylabel = 'Frequency')
```

```python
# Now, Let's check what we will get if we increase the range from 0 to 100 and also increase the size to 10,000
random_check = [int(np.random.normal(50, 12.75)) for i in range(10000)]
values, frequencies = np.unique(random_check, return_counts = True)
```

```python
# Another method to plot the histogram and we can see that chart is more smoother if we increase the range 
# to 100 and size to 10,000
sns.set(color_codes = True)
ax = sns.distplot(random_check,
                  bins = 100,
                  kde = True,
                  color = 'skyblue',
                  hist_kws = {"linewidth": 15,'alpha':1})
ax.set(xlabel = 'Values', ylabel = 'Frequency')
ax.set_title('Normal Distribution')
```

```python
# Below Line plot is to represent the value variation in the randomly generated distribution numbers
axes = sns.lineplot(color = "coral", data = np.array(random_check))
axes.set(xlabel = 'Normal Distribution Number Index', ylabel = 'Value', title = 'Normal Values - Line Plot')
```

```python
# Below Line plot is to represent the normal distribution b/w Frequency vs Randomly generated numbers
axes= sns.lineplot(color = "blue", x = values, y = frequencies)
axes.set(xlabel='Values', ylabel = 'Frequency', title = 'Normal Distribution - Line Plot')
```

# Summary

We used Random module, Numpy and Seaborn for the above data analysis. We have shown that we can get normally distributed random numbers by use of random module and Numpy module. The Mean value is 12.5 and the standard deviation is 3.88.
We explored various methods to draw Normal Distribution charts such as bar graphs and line graphs for the range 0 to 25 and 0 to 100. We observed that as the sample size increases to 10000, most of the values are closer to the mean and have small variance and hence the bar charts are more smoother.
