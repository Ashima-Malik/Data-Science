```python
import numpy as np
import random
import seaborn as sns
random.seed(89)
```

```python
# Here, We have used random module to genertae uniformly distributed random values
# Here, we have used range from 0 to 26 to get the values between 0 and 25 inclusive 
uniform_values = [int(random.uniform(0, 26)) for i in range(1000)]
```

```python
# Mean of uniform values
print("The mean of uniform normal distribution is:", np.mean(uniform_values))
```

```python
# Here, this is the another method to get uniformly distributed random values using Numpy. 
# numpy.random.uniform(low=0.0, high=26.0, size=1000)
uniform_values_np = np.random.uniform(0, 26, 1000)
```

```python
# Here, values and frequencies unique count obtained from the uniform_values object to use these in seaborn plotting
values, frequencies = np.unique(uniform_values, return_counts = True)
```

```python
print(values)
```

```python
print(frequencies)
```

```python
sns.set_style('whitegrid')
axes = sns.barplot(x = values, y = frequencies, color = 'gray')
axes.set_title('Uniform Distribution')
axes.set(xlabel = 'values', ylabel = 'Frequency')
```

```python
# Below Line plot is to represent the uniform distribution b/w Frequency vs Randomly generated numbers for range 0 to 25
axes = sns.lineplot(color = "blue", x = values, y = frequencies)
axes.set(xlabel = 'Values', ylabel = 'Frequency', title = 'Uniform Distribution - Line Plot')
axes.set_ylim(0,)
```

```python
# Now, Let's check what we will get if we increase the range and also increase the size to 10,000
check = np.floor(np.random.uniform(0, 101, 10000))
values_check, frequencies_check = np.unique(check, return_counts = True)
```

```python
# Another method to plot the histogram and we can see that chart is more smoother if we increase the range 
# to 100 and size to 10,000
sns.set(color_codes = True)
ax = sns.distplot(check,
                  bins = 100,
                  kde = True,
                  color = 'skyblue',
                  hist_kws = {"linewidth": 15,'alpha':1})
ax.set(xlabel = 'Values', ylabel = 'Frequency')
ax.set_title('Uniform Distribution')
```

```python
# Below Line plot to represent the value variation in the randomly generated distribution numbers
check = np.floor(np.random.uniform(0, 101, 10000))
axes = sns.lineplot(color = "coral", data = check)
axes.set(xlabel = 'Uniform Number Index', ylabel = 'Value', title = 'Uniform Values - Line Plot')
```

```python
# Below Line plot is to represent the uniform distribution b/w Frequency vs Randomly generated numbers for range 0 to 100
axes= sns.lineplot(color = "blue", x = values_check, y = frequencies_check)
axes.set(xlabel = 'Values', ylabel = 'Frequency', title = 'Uniform Distribution - Line Plot')
axes.set_ylim(0,)
```
