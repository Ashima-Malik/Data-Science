```python
import numpy as np
import random
import seaborn as sns
```

```python
#Here, we have calculated exponentially distributed values using random package. The thought process behind it was
# that we had to accumulate 1000 randomly generated exponential values mainly in the range of 0 to 25. 
#Some of the extreme values will be outside the range. 

lst_exp = [int(random.expovariate(1/12.5)) for i in range(1000)]
values, frequencies = np.unique(lst_exp,return_counts = True) 
```

```python
print(values)
```

```python
print(frequencies)
```

```python
sns.set_style('whitegrid')
axes = sns.barplot(x = values, y = frequencies, color = 'blue')
axes.set_title('Exponential Distribution')
axes.set(xlabel = 'Number', ylabel ='Frequency', xticks = range(0,78,3)) # 78 and 3 are set to keep the data in range(0,25)
print('Limited Random Integer Values')
```

```python
values, frequencies = np.unique(lst_exp, return_counts = True)
axes= sns.lineplot(color = "blue", x = values, y = frequencies)
axes.set(xlabel = 'Values', ylabel = 'Frequency', title = 'Exponential Distribution - Line Plot', xticks = range(0,28,5))
print('')
```

```python
# Now, Let's check what we will get if we increase the range from 0 to 100 and also increase the size to 10,000
lst_exp_check = []
mean= 12.5
lambd = 1.0/mean
for i in range(10000):    
    num = int(random.expovariate(lambd))
    lst_exp_check.append(num)

print(len(lst_exp_check))
```

```python
# Another method to plot the distogram and we can see that chart is more smoother if we increase the range 
# to 100 and size to 10,000

sns.set(color_codes = True)
ax = sns.distplot(lst_exp_check,
                  bins = 100,
                  kde = True,
                  color = 'skyblue',
                  hist_kws = {"linewidth": 15,'alpha':1})
ax.set(xlabel = 'Values', ylabel = 'Frequency')
ax.set_title('Exponential Distribution')
```

```python
# Below Line plot is to represent the value variation in the randomly generated distribution numbers
axes = sns.lineplot(color = "coral", data = np.array(lst_exp_check))
axes.set(xlabel = 'Exponential Distribution Number Index', ylabel = 'Value', title = 'Exponential Values - Line Plot')
```

```python
# Below Line plot is to represent the exponential distribution b/w Frequency vs Randomly generated numbers
values, frequencies = np.unique(lst_exp_check, return_counts=True)
axes= sns.lineplot(color = "blue", x = values, y = frequencies)
axes.set(xlabel = 'Values', ylabel = 'Frequency', title = 'Exponential Distribution - Line Plot')
print('Charts would be smoother if increased the range.')
```

# Summary

We used Random module, Numpy and Seaborn for the above data analysis. We have shown that we can get exponentially distributed random numbers by use of random module and Numpy module. We generated 1000 randomly generated exponential values mainly in the range of 0 to 25. Then, we genertaed 10000 exponentially distributed values in range 0 to 100.
We explored various methods to draw Histrogram such as sns.barplot and sns.distplot. We also tried to make line charts for both the ranges.
By comparing the charts with different range,we can observe that the on increasing the sample size the distribution of the means becomes more pointy and narrow, indicating that the true mean becomes more and more accurate. Hence, the bar chart and line chart are more smoother for higher sample size(10,000) compared to small sample size(1000).
