```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

import seaborn as sns

plt.style.use('fivethirtyeight')

%matplotlib inline
%config InlineBackend.figure_format = 'retina'

# Import LogisticRegression and LinearRegression from sklearn.linear_model
from sklearn.linear_model import LogisticRegression, LinearRegression
```

```python
url = 'https://raw.githubusercontent.com/persocom01/GA-lessons/master/4.01-lesson-logistic_regression/datasets/admissions.csv'
df = pd.read_csv(url)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace = True)
```

```python
df.groupby('admit')['prestige'].mean()
```

```python
admit = df[df['prestige']== 1.0]
admit
```

```python
admit = pd.concat([admit] * 10, axis = 0)
```

```python
admit
```

```python
np.random.seed(42)
```

```python
admit.loc[admit['admit']==1, 'gpa'] += np.random.random(size = admit[admit.admit==1].shape[0])
admit.loc[admit['admit']==0, 'gpa'] -= np.random.random(size = admit[admit.admit==0].shape[0])
```

```python
y = admit['admit']
X = admit[['gpa']]
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 42)
```

```python
lr = LogisticRegression()
lr.fit(X_train, y_train)
```

```python
lr.intercept_
```

```python
lr.coef_
```

```python
lr.predict(X_train)
```

```python
lr.predict_proba(X_train)
```
